```python
def find_faces(image, scale=1.1, minNeighbors=5, cascade = "haarcascade_frontalface_default.xml"):

    faceCascade = cv2.CascadeClassifier(cascade)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=scale,
        minNeighbors=minNeighbors,
        minSize=(30, 30),
        #flags = cv2.cv.CV_HAAR_SCALE_IMAGE
    )

    print("Funnet {0}".format(len(faces)))
    return faces
```

```python
def mark_faces(faces, image):
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
```

```python
nb_faces = find_faces(nb_image, scale=1.01, minNeighbors=5)
nb_faces
```
